# Laços de Repetição

>É comum encontrar uma situação em que comandos precisam ser repetidos várias vezes. Quando isso acontece são usados loops (ou laços) que permitem repetir comandos por uma dada quantidade de vezes ou enquanto uma condição seja satisfeita.

>Vamos ver dois tipos de loop, o `while` e o `for`.

>O `while` executa instruções enquanto um condição é atendida.

>O `for` percorre os elementos de uma sequência e, para cada item, executa o bloco de código.

# Sentença `while`

>O laço de repetição `while` primeiro verifica a condição, se ela for verdadeira, o bloco de código será executado. Depois que os comandos são executados, a condição é verificada novamente e tudo irá se repetir enquanto a condição for verdadeira.

```python
valor = 0

print("Começo do laço de repetição")

while valor < 10 :   # Enquanto valor for menor que 10 faça:
    print(valor)    # Imprima valor
    valor += 1   # Adicione 1 ao valor (valo = valor + 1)

print("Fim do laço de repetição")
```


>Assim como na sentença `if`, para o `while` também existe a sentença `else`.

>Da mesma forma, enquanto a condição for verdadeira o bloco de código do `while` é repetido, caso contrario, o bloco de código do `else` é executado.

```python
y = 0
while y <= 10:
  y += 1
else:
  print("O valor de y é:", y)
```

**Break**

>No `while` só ocorre a verificação da condição uma vez no inicio de cada repetição. Porém, às vezes, é necessário fazer alguma validação dentro do `while`. Para isso podemos usar o `break` que interrompe o laço de repetição.

>Exemplo:

```python
valor = 0

while True:                                         # Loop infinito
  valor = int(input("Insira um valor positivo: "))
  if valor < 0:                                     # Se o valor for menor que 0
    break                                           # Sai do loop

print("Acabou o laço de repetição")
```

**Continue**

>Considerando uma situação em que é preciso retornar ao início do laço de repetição sem terminar de executar todo o bloco de código, para realização essa ação será usado o `continue`.

>Exemplo:



```python
i = 0

while i < 20 :
  i += 1
  if i % 2 == 0:  # Se "i" for par
    continue      # Volte para o começo do laço
  print(i)
```

# Sentença `for`

>No loop `for` passamos um contador (ou iterador) sobre itens de uma sequência finita (listas ou strings, por exemplo), seguindo a ordem que aparece na sequência. Exemplo:

```python
lista = [0,1,2,3,8,4,6,4]

for x in lista: # Para x no intervalo dos valores dos itens de lista
  print(x) # Imprima x
```

>Similar ao `while`, na sentença `for` também existe a sentença `else`.

>Então, enquanto a sequência do `for` é iterada, bloco de código é repetido. O bloco de código do `else` é executado depois do final do loop, se não tiver o `break`.

>Exemplo:

```python
lista = [1,2,3,4,5]

for x in lista:
  print(x)
else:
  print("Fim do laço")
```

**Break**

>Realiza a mesma ação que para o `while`, isto é, usamos o `break` para interromper o laço.

>Exemplo:

```python
lista = [1,2,3,4,5,6,7]
for num in lista:
  if num == 10:  # Se o número for igual ao 3
    break       # Sai do loop
  print(num)
```

**Continue**

>Assim como no `while`, passa para a próxima iteração (próximo valor da lista).

>Exemplo:

```python
numeros = [1,2,3,4,5,6,7]
for x in numeros:
  if x % 2 == 0:  # Se x for par
    continue      # Passa para a próxima iteração, imprimindo a lista sem os pares
  print(x)
```

**Função range**

>A função `range` pode ser usada para construir a lista de iteração do loop `for` e pode receber o início e o fim, ou também receber 3 parâmetros: o início do laço, o fim e o quanto ele deve acrescentar no contador por cada iteração:

>`range(início, final, incremento)`

>`início`: opcional, por padrão é zero.

>`parada`: é o último.

>`incremento`: opcinal, valor acrescentado em cada iteração, por padrão é 1.

>Exemplo

```python
for x in range(5):
  print(x)
```

```python
for num in range(2,7):
  print(num)
```

```python
for valor in range(3,30,3):
  print(valor)
```
