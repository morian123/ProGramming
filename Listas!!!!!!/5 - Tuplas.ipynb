# Lesson 05: Tuplas


# Introdução


>São estruturas de dados parecidas com listas, mas com a particularidade de serem **imutáveis.**

>Tuplas são sequências e, assim como listas, podem ser indexadas e fatiadas, mas **não é possível modificá-las.**

>Um valor do tipo tupla é uma série de valores separados por vírgulas e entre parênteses.


```python
x = (1,2,3)
print(x)
```

```python
x[0]=1
```

# A função **tuple**

>A função tuple constrói uma tupla a partir de uma sequência qualquer

```python
tuple("abcdefg")
```

```python
tuple([1,2,3])
```

# Formatando tuplas

>Um **template** pode ser aplicado aos diversos valores de uma tupla para construir uma string formatada

```python
template = ("%s tem %d anos")
tupla = ('Pedro',10)
print(template % tupla)
```

#Atribuição de tuplas

> De vez em quando, é necessário **trocar entre si os valores de duas variáveis**.Com operações de atribuição convencionais, temos que utilizar uma variável temporária. Python fornece uma **forma de atribuição de tupla** que resolve esse problema elegantemente:


```python
a=1000
b=1
temp = a
a = b
b = temp
print("Valor de a: ", a)
print("Valor de b: ", b)
```

```python
a=1000
b=1
a , b = b , a
print("Valor de a: ", a)
print("Valor de b: ", b)
```

# Especificações de formato

```python
from math import pi
print("Ponto flutuante: %f" % pi)
print("Ponto flutuante com 20 casas decimais: ||%.20f||" % pi)
print("Ponto flutuante com 10 caracteres: ||%10f||" % pi )
print("Ponto flutuante com 10 caracteres e 2 casas decimais: ||%10.2f||" % pi )
print("Ponto flutuante em notacao cientifica: ||%10e||" % pi )
```
