> A f√≥rmula para calcular a √°rea de uma circunfer√™ncia √©: area = œÄ * raio¬≤. Considerando para este problema que œÄ = 3.14159:
>
- Efetue o c√°lculo da √°rea, elevando o valor de raio ao quadrado e multiplicando por œÄ.
>
> **Entrada**
>
> A entrada cont√©m um valor de ponto flutuante (float), no caso, a vari√°vel raio.
>
> **Sa√≠da**
>
> Apresentar a mensagem "A=" seguido pelo valor da vari√°vel area, conforme exemplo abaixo, com 4 casas ap√≥s o ponto decimal. Utilize vari√°veis float. Como todos os problemas, n√£o esque√ßa de imprimir o fim de linha ap√≥s o resultado, caso contr√°rio, voc√™ receber√° "Presentation Error".

# Exerc√≠cio 3


> Leia 3 valores de ponto flutuante e efetue o c√°lculo das ra√≠zes da equa√ß√£o de Bhaskara. Se n√£o for poss√≠vel calcular as ra√≠zes, mostre a mensagem correspondente ‚ÄúImpossivel calcular‚Äù, caso haja uma divis√£o por 0 ou raiz de numero negativo.
>
> **Entrada**
>
> Leia tr√™s valores de ponto flutuante (float) A, B e C.
>
> **Sa√≠da**
>
> Se n√£o houver possibilidade de calcular as ra√≠zes, apresente a mensagem "Impossivel calcular". Caso contr√°rio, imprima o resultado das ra√≠zes com 5 d√≠gitos ap√≥s o ponto, com uma mensagem correspondente conforme exemplo abaixo. Imprima sempre o final de linha ap√≥s cada mensagem.

```python
# Escreva a solu√ß√£o aqui!
# Dica: utilize o comando .split() para separar os itens do input() e depois os
# converta para float. Exemplo abaixo:

A, B, C = input().split()
A = float(A)
B = float(B)
C = float(C)

delta = (B**2) - 4*A*C

if delta < 0:
  print("Impossivel calcular")
elif delta == 0:
  x = (- B + 0)/(2*A)
  print(f'X = {x}')
else:
  x1 = (- B + (delta**0.5))/(2*A)
  x2 = (- B - (delta**0.5))/(2*A)
  print(f'X1 = {x1}\nX2 = {x2}')




```

# Exerc√≠cio 5

> Suponha que voc√™ est√° desenvolvendo um jogo 2D de tiro ao alvo e voc√™ deseja verificar se um tiro acertou o alvo. Assuma que o tiro e o alvo s√£o c√≠rculos e s√£o conhecidas suas posi√ß√µes e raios.
>
> O programa deve receber as coordenadas dos centros dos c√≠rculos (x,y) e seus raios (r) na mesma linha, respectivamente. Exemplo de entrada:
>
- 10.5  12  3 (x,y e r do alvo)
-  14.7  13.2  8 (x,y e r do tiro)



> <a href="https://peteletrica.ufes.br/petcode/">

```python
# Escreva a solu√ß√£o aqui!
# Dica: utilize o comando .split() para separar os itens do input() e depois os
# converta para float. Exemplo abaixo:
# Dica 2: para que haja uma colis√£o, a dist√¢ncia entre os centros dos c√≠rculos
# (alvo e tiro) deve ser menor que os raios somados!

x_alvo, y_alvo, r_alvo = input().split()
x_alvo = float(x_alvo)
y_alvo = float(y_alvo)
r_alvo = float(r_alvo)

x_tiro, y_tiro, r_tiro = input().split()
x_tiro = float(x_tiro)
y_tiro = float(y_tiro)
r_tiro = float(r_tiro)

distancia = (((x_alvo-x_tiro)**2) + ((y_alvo-y_tiro)**2))**(0.5) # formula para
                                                                 # distancia entre dois pontos
if distancia < (r_alvo + r_tiro):
  print('Colis√£o detectada.')
else:
  print('Colis√£o n√£o detectada.')






```

> Considere um jogo que tenha um tanque com
um canh√£o que se move ao em rela√ß√£o ao eixo

do tanque (ty, tx) e seu tamanho (h, w), al√©m
do √¢ngulo do canh√£o ùúÉ e seu tamanho t.
Mostre na tela a posi√ß√£o da base do canh√£o
(by, bx), assumindo que ela est√° na metade da
largura do tanque, e a posi√ß√£o da ponta do

>
> <a href="https://peteletrica.ufes.br/petcode/">

```python
# Escreva a solu√ß√£o aqui!
# Dica: utilize rela√ß√µes trigonom√©tricas com a biblioteca math

import math

ty, tx = input("Informe a posi√ß√£o do tanque (ty,tx): ").split()
ty = float(ty)
tx = float(tx)

h, w = input("Informe o tamanho do tanque (h,w): ").split()
h = float(h)
w = float(w)

ang, t = input("Informe o √¢ngulo e tamanho do canh√£o (ang,t): ").split()
ang =  math.radians(float(ang))
t = float(t)

by = ty
bx = tx + (w/2)

cy = by - (t*math.cos(ang))
cx = bx + (t*math.sin(ang))


print(f"Base do canh√£o (by, bx): ({by}, {bx})")
print(f"Ponta do canh√£o (cy, cx): ({cy}, {cx})")
```
